# Filter

```python
from scipy.signal import butter, lfilter

def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a
def butter_bandpass_filter(signal, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    n_trials, n_chans, n_time = signal.shape
    filtered = np.zeros(signal.shape)
    print(filtered.shape)
    for i in range(n_trials):
        for j in range(n_chans):
            filtered[i][j] = lfilter(b, a, signal[i][j])
    return filtered


data_train = butter_bandpass_filter(data, 0.2,13,2500, order=5)
data_test = butter_bandpass_filter(tdata, 0.2,13,2500, order=5)
print('filtered train data shape', data_train.shape)
print('filtered test data shape', data_test.shape)
```

# Covariance Matrix , Channel Selection , Vectorization  , Classifier

```python
from pyriemann.estimation import Covariances
from pyriemann.tangentspace import TangentSpace
from pyriemann.channelselection import ElectrodeSelection
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.metrics import classification_report, auc, roc_auc_score
import pandas as pd
from sklearn import mixture
from sklearn.multiclass import OneVsRestClassifier
from sklearn.multiclass import OneVsOneClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import shuffle

# number of features to select 
nelec =  [20,28,40]  #[34,36,38,40]      # [20,22,24,26,28,30,32]      
for ii in nelec:
    print(ii*(ii+1)/2)
    #########################  calculating covariance matrix of train data ##########################################
    covmat = Covariances(estimator='lwf').fit(training_data, training_labels)
    cov_train = covmat.transform(training_data)
    # feature selection 
    channel_train = ElectrodeSelection(nelec =ii).fit(cov_train, training_labels, sample_weight=None)
    select_train = channel_train.transform(cov_train)
    
    # Vectorization of upper triangular matrix 
    s1 = [[]*3456 for k in range(3456)]
    #s1=np.array([1])
    p3=np.empty((0,int(ii*(ii+1)*1/2)))
    for dim in range(3456):
        b = np.triu(select_train, k=0)
        #print('b_shape=', b.shape)
        #b = np.transpose(b)
        for i in range(0,ii):
           
            g1 = b[dim][i]
            g1 = np.transpose(g1)
            s =  [x for x in list((g1)) if x!=0]  
            #print('s _shape=', np.shape(s[:-1]))
            for j,elem in enumerate(s[:-1]):
              
                elem1 =  np.sqrt(2)*elem
                s1[dim].append(elem1)
               
            s1[dim].append(s[-1])
         #   print('s shape =', np.shape(s1[dim]))
       # print('final_ shape =', np.shape(s1[dim]))
        f1 = np.reshape(s1[dim],(1,int(ii*(ii+1)*1/2)))
        p3 = np.concatenate((p3,f1),axis=0)
    print('p3=',np.shape(p3))
    p3[np.isnan(p3)]=0
        #for dim in range(3456):
           
                


    ######################## validation data #########################
    cov_val = covmat.transform(validation_data)
    select_val = channel_train.transform(cov_val)
    
    s2 = [[]*864 for k in range(864)]
   
    p4=np.empty((0,int(ii*(ii+1)*1/2)))
    for dim1 in range(864):
        b1 = np.triu(select_val, k=0)
       # b1 = np.transpose(b1)
        #print('b_shape=', b.shape)
        for i in range(0,ii):
            g2 = b1[dim1][i]
            g2 = np.transpose(g2)
          
            s3 =  [x for x in list((g2)) if x!=0]  
            #print('s _shape=', np.shape(s[:-1]))
            for j1,elem2 in enumerate(s3[:-1]):
              
                elem3 =  np.sqrt(2)*elem2
                s2[dim1].append(elem3)
               # print('fffff=', np.shape(s1))
            s2[dim1].append(s3[-1])
           # print('s shape =', np.shape(s1))
        #print('final_ shape =', np.shape(s1))
        f2 = np.reshape(s2[dim1],(1,int(ii*(ii+1)*1/2)))
        p4 = np.concatenate((p4,f2),axis=0)
    print('p4=',np.shape(p4))
    p4[np.isnan(p4)]=0
      

   
    ##################################test data  ######################################
    cov_test = covmat.transform(test_data)
    select_test = channel_train.transform(cov_test)
    
    s4 = [[]*1230 for k in range(1230)]
    
    p5=np.empty((0,int(ii*(ii+1)*1/2)))
    for dim2 in range(1230):
        b2 = np.triu(select_test, k=0)
      #  b2 = np.transpose(b2)
        #print('b_shape=', b.shape)
        for i in range(0,ii):
            #print('iiiiiiiiiiiiiiiiiiiiiiii=', i)
            g3 = b2[dim2][i]
            g3 = np.transpose(g3)
            s5 =  [x for x in list((g3)) if x!=0]  
            #print('s _shape=', np.shape(s[:-1]))
            for j2,elem4 in enumerate(s5[:-1]):
            
                elem5 =  np.sqrt(2)*elem4
                s4[dim2].append(elem5)
           
            s4[dim2].append(s5[-1])
           # print('s shape =', np.shape(s1))
       # print('final_ shape =', np.shape(s1))
        f3 = np.reshape(s4[dim2],(1,int(ii*(ii+1)*1/2)))
        p5 = np.concatenate((p5,f3),axis=0)
    print('p5=',np.shape(p5))
    p5[np.isnan(p5)]=0
        #for dim in range(3456):
    
    
    ########################## Multi-layer Perceptron (MLP) Classifier #######################################
    #classifier = SVC(probability=True)
    classifier = MLPClassifier(max_iter=10000,activation = 'relu',solver='adam',random_state=1)  
    #OneVsRestClassifier(LinearSVC(max_iter=20000,random_state=0))
    #MLPClassifier(hidden_layer_sizes=(150,100,50), max_iter=300,activation = 'relu',solver='adam',random_state=1)
    classifier.fit(p3, training_labels)

    training_predicted= classifier.predict(p3)
    print('training_labels =',list(training_predicted).count(0), list(training_predicted).count(1), list(training_predicted).count(2))
    validation_predicted=classifier.predict(p4)
    print('validation_labels =',list(validation_predicted).count(0), list(validation_predicted).count(1), list(validation_predicted).count(2))
    test_predicted=classifier.predict(p5)
    print('test_labels =',list(test_predicted).count(0), list(test_predicted).count(1), list(test_predicted).count(2))
    dt1 = pd.DataFrame(training_predicted).to_csv('/media/ubuntu/HDD_P1/shubh@/new3/tlab1_{}.csv'.format(ii))
    dt2 = pd.DataFrame(validation_predicted).to_csv('/media/ubuntu/HDD_P1/shubh@/new3/vlab1_{}.csv'.format(ii))
    dt3 = pd.DataFrame(test_predicted).to_csv('/media/ubuntu/HDD_P1/shubh@/new3/test_lab1_{}.csv'.format(ii))
    
    ########################### OneVsRest SVM Classifier #######################################
    from sklearn.svm import LinearSVC, SVC 
    classifier1 = OneVsRestClassifier(LinearSVC(max_iter=1000,random_state=0))
    #SVC(gamma='scale', class_weight='balanced')
    classifier1.fit(p3, training_labels)

    train_pred= classifier1.predict(p3)
    val_pred=classifier1.predict(p4)
    test_pred=classifier1.predict(p5)
    print('train_svm_labels =',list(train_pred).count(0), list(train_pred).count(1), list(train_pred).count(2))
    
    print('val_svm_labels =',list(val_pred).count(0), list(val_pred).count(1), list(val_pred).count(2))
    
    print('test_svm_labels =',list(test_pred).count(0), list(test_pred).count(1), list(test_pred).count(2))
    dt4 = pd.DataFrame(train_pred).to_csv('/media/ubuntu/HDD_P1/shubh@/new3/tlab1_svm{}.csv'.format(ii))
    dt5 = pd.DataFrame(val_pred).to_csv('/media/ubuntu/HDD_P1/shubh@/new3/vlab1_svm{}.csv'.format(ii))
    dt6 = pd.DataFrame(test_pred).to_csv('/media/ubuntu/HDD_P1/shubh@/new3/test1_svmlab_s{}.csv'.format(ii))

   

    #Importing Confusion Matrix to calculate accuracy score 

    from sklearn.metrics import confusion_matrix
    

    cm1 = confusion_matrix(training_predicted, training_labels)
    cm2 = confusion_matrix(validation_predicted, validation_labels)
    cm3 = confusion_matrix(train_pred, training_labels)
    cm4 = confusion_matrix(val_pred, validation_labels)
    
    #Printing the accuracy
    print('Accuracy of MLPClassifier :', accuracy(cm1),accuracy(cm2))           
    print('Accuracy of SVMClassifier :', accuracy(cm3),accuracy(cm4))   #,accuracy(cm5),accuracy(cm6))
```
